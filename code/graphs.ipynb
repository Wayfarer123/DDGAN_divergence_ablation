```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style='darkgrid')
```

```python
with open('FID_scores_FMNIST_ddgan_h2.npy', "rb") as f:
    scores_h2 = np.load(f)
with open('FID_scores_FMNIST_ddgan_rkl.npy', "rb") as f:
    scores_rkl = np.load(f)
with open('FID_scores_FMNIST_ddgan_tv.npy', "rb") as f:
    scores_tv = np.load(f)
with open('FID_scores_FMNIST_ddgan.npy', "rb") as f:
    scores = np.load(f)
with open('FID_scores_FMNIST_ddgan_wd.npy', "rb") as f:
    scores_wd = np.load(f)
```

```python
import seaborn as sns
sns.set_theme(style='darkgrid')

plt.figure(figsize=(12,8))

num_steps = np.arange(1, 11, 1)

plt.plot(num_steps, scores, lw=3, label='JS (basic GAN)')
plt.plot(num_steps, scores_h2, lw=3, label='H$^2$')
plt.plot(num_steps, scores_rkl, lw=3, label='Reversed-KL')
plt.plot(num_steps, scores_tv, lw=3, label='Total variation')
plt.plot(num_steps, scores_wd, lw=3, label='Wasserstein')

plt.xlabel("T", fontsize=18)
plt.ylabel("FID", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(left=0.5, right=10.5)

plt.legend(loc='upper right', fontsize=18, shadow=True)

plt.savefig("DDGAN_ALL_FID_FMNIST.png", bbox_inches='tight')
plt.show()
```
